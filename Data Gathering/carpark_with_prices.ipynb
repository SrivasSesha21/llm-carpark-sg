```python
# Function to calculate distance
def calculate_distance(row, rate_row):
    coords_1 = (row['Latitude'], row['Longitude'])
    coords_2 = (rate_row['latitude'], rate_row['longitude'])
    return haversine(coords_1, coords_2, unit = Unit.MILES)

def find_closest_carpark(row, df_rates):
    distances = df_rates.apply(lambda rate_row: calculate_distance(row, rate_row), axis=1)
    closest_index = distances.idxmin()
    return df_rates.loc[closest_index]

closest_carparks = df_carpark.apply(lambda row: find_closest_carpark(row, df_rates), axis=1)
df_carpark['Price_Weekdays_before_5_6pm'] = closest_carparks['Weekdays  before 5/6pm']
df_carpark['Price_Weekdays_after_5_6pm'] = closest_carparks['Weekdays  after 5/6pm']
df_carpark['Price_Saturdays'] = closest_carparks['Saturdays']
df_carpark['Price_Sundays_Public_Holiday'] = closest_carparks['Sundays/Public Holidays']
```
